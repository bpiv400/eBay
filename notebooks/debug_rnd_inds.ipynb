```python
import numpy as np
import pandas as pd
import os
import sys
import math
import datetime
import gc
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
binned_df = pd.read_csv('data/exps/rnn1/binned/toy.csv')
feats_df = pd.read_csv('data/toy/toy-1_feats2.csv')
org_df = pd.read_csv('data/toy/toy-1.csv')
prep_df = pd.read_csv('data/exps/rnn1/toy-1.csv')
```

```python
def grab_col(prep_df, binned_df, colname):
    # extract the offr column for b3 from binned
    # and extract the ref offer fro b3 from binned
    binned = binned_df[colname].values
    binned = binned[~np.isnan(binned)]
    
    prepped = prep_df[colname].values
    prepped = prepped[~np.isnan(prepped)]
    
    return prepped, binned

def plots(prepped, binned, colname):
    hist = plt.hist(x=binned)
    plt.title("Binned Distribution of %s" % colname)
    plt.show()
    
    hist = plt.hist(x=prepped)
    plt.title("Prepped Distribution of %s" % colname)
    plt.show()
def wrapper(prep_df, binned_df, colname):
    prepped, binned = grab_col(prep_df, binned_df, colname)
    plots(prepped, binned, colname)
    
```

```python
# extract the offr column for b3 from binned
# and extract the ref offer fro b3 from binned
normed_bin = binned_df['offr_b3'].values
normed_bin = normed_bin[~np.isnan(normed_bin)]
ref_bin = binned_df['ref_offr_b3'].values
ref_bin = ref_bin[~np.isnan(ref_bin)]

normed_bin = binned_df['offr_b3'].values
normed_bin = normed_bin[~np.isnan(normed_bin)]
ref_bin = binned_df['ref_offr_b3'].values
ref_bin = ref_bin[~np.isnan(ref_bin)]
```
